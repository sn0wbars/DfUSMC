```python
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import ipywidgets as w
```

```python
testPath = os.path.join('Dataset', 'Bikes.avi')
```

```python
max_num = 30
def loadClip(fullpath=testPath):
    cap = cv2.VideoCapture(fullpath);
    if(not cap.isOpened()):
        return
    try:
        num_image=int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
#         print(num_image)
        num_image = max_num if (num_image>max_num) else num_image; #only first 30
        imgs =[cap.read()[1][:,:,::-1] for _ in range(num_image)]
        return imgs
    finally:
        cap.release()

#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
```

```python
imgs = loadClip()
# plt.imshow(imgs[0])
grays = [cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) for img in imgs]
# plt.imshow(grays[0], cmap='gray')
```

```python
ref = grays[0]
ref_features = cv2.goodFeaturesToTrack(ref, 20000, 1e-10, 5)
```

```python
ref_features[:,0,0]
```

```python
plt.imshow(imgs[0])
step = 10
plt.scatter(ref_features[::step,0,0], ref_features[::step,0,1], s=1, c='r')
```

```python
def plot_with_keypoints(i):
    plt.imshow(imgs[i])
    
    features = cv2.goodFeaturesToTrack(grays[i], 20000, 1e-10, 5)    
    step = 10
    plt.scatter(features[::step,0,0], features[::step,0,1], s=1, c='r')
    plt.axis('off')
_ = w.interact(plot_with_keypoints, i=w.IntSlider(0,0,len(grays)-1))
```
